```python
import zipfile
import pandas as pd
import numpy as np

import pandas as pd
import numpy as np

# Adjust paths to the location of your extracted CSV files
flights_data = pd.read_csv(r'C:\Users\soumy\OneDrive\Desktop\assignments\Int to Prgm\data/flights.csv')
weather_data_pd = pd.read_csv(r'C:\Users\soumy\OneDrive\Desktop\assignments\Int to Prgm\data/weather.csv')

# Convert weather data to numpy array for questions requiring numpy
weather_data_np = weather_data_pd.to_numpy()

# Display a sample of the data to confirm
print(flights_data.head())
print(weather_data_pd.head())


#Question 1 How many flights were there from JFK to SLC? Int
print()
print('1', 'output')
#Filtering the flights_data DataFrame to include only flights from JFK to SLC
#Using 'shape[0]' to count the number of rows (flights) that match this condition
q_1 = flights_data[(flights_data['origin'] == 'JFK') & (flights_data['dest'] == 'SLC')].shape[0]
print("Total flights from JFK to SLC:", q_1)


#Question 2 How many airlines fly to SLC? Should be int
print()
print('2', 'output')
#Selecting flights with destination SLC and counting unique carriers (airlines)
# The 'nunique()' function returns the count of distinct values in the 'carrier' column
q_2 = flights_data[flights_data['dest'] == 'SLC']['carrier'].nunique()
print("Number of airlines flying to SLC:", q_2)

# Question 3 What is the average arrival delay for flights to RDU? float
print()
print('3', 'output')
#Calculating the average arrival delay for flights destined for RDU
# 'mean()' computes the average for the 'arr_delay' column after filtering by destination 'RDU'
q_3 = flights_data[flights_data['dest'] == 'RDU']['arr_delay'].mean()
print("Average arrival delay for flights to RDU:", q_3)

# Question 4 What proportion of flights to SEA come from the two NYC airports (LGA and JFK)?  float
print()
print('4', 'output')
#Calculating the proportion of flights to SEA originating from either LGA or JFK
# We count total flights to SEA and flights to SEA from LGA or JFK, then divide
total_sea_flights = flights_data[flights_data['dest'] == 'SEA'].shape[0]  # Total flights to SEA
nyc_sea_flights = flights_data[(flights_data['dest'] == 'SEA') & 
                               (flights_data['origin'].isin(['LGA', 'JFK']))].shape[0]  # NYC flights to SEA
q_4 = nyc_sea_flights / total_sea_flights if total_sea_flights > 0 else 0  # Avoid division by zero
print("Proportion of flights to SEA from NYC airports:", q_4)


# Question 5 Which date has the largest average depature delay? Pd slice with date and float
print()
print('5', 'output')
#Converting year, month, and day columns into a single 'date' column
# Using 'groupby' to find the average departure delay for each date, and then finding the max
flights_data['date'] = pd.to_datetime(flights_data[['year', 'month', 'day']])  # Combine to datetime
average_dep_delay = flights_data.groupby('date')['dep_delay'].mean()  # Group and calculate mean delay
q_5 = average_dep_delay.idxmax(), average_dep_delay.max()  # Find the date with max average delay
print("Date with largest average departure delay:", q_5)

#Question 6 Which date has the largest average arrival delay? pd slice with date and float
print()
print('6', 'output')
#Calculating based on arrival delay instead of departure delay
average_arr_delay = flights_data.groupby('date')['arr_delay'].mean()  # Mean arrival delay by date
q_6 = average_arr_delay.idxmax(), average_arr_delay.max()  # Date with the highest arrival delay
print("Date with largest average arrival delay:", q_6)

#Question 7 Which flight departing LGA or JFK in 2013 flew the fastest? pd slice with tailnumber and speed
#speed = distance/airtime
print()
print('7', 'output')
# Calculate speed = distance / air_time (converted to hours) and 
#find the fastest flight for LGA or JFK in 2013
flights_data['speed'] = flights_data['distance'] / (flights_data['air_time'] / 60)  # Convert air_time to hours
fastest_flight = flights_data[(flights_data['origin'].isin(['LGA', 'JFK'])) & (flights_data['year'] == 2013)]
fastest_flight = fastest_flight.loc[fastest_flight['speed'].idxmax()]  # Find the row with max speed
q_7 = fastest_flight[['tailnum', 'speed']]
print("Fastest flight departing from LGA or JFK in 2013:", q_7)


# Question 8 Replace all nans in the weather pd dataframe with 0s. Pd with no nans
print()
print('8', 'output')
#Using 'fillna(0)' to replace all NaN values with 0
# This operation modifies the DataFrame in place (or returns a modified copy if assigned)
q_8 = weather_data_pd.fillna(0)
print("Weather data with NaNs replaced by 0s:\n", q_8.head())

# Question 9 How many observations were made in Feburary? Int
print()
print('9', 'output')
#Counting rows where the 'month' column equals 2 (February)
q_9 = weather_data_pd[weather_data_pd['month'] == 2].shape[0]
print("Number of observations in February:", q_9)

#Question 10 What was the mean for humidity in February? Float
print()
print('10','output')
# Filter for February and calculate the mean of the 'humidity' column, ignoring NaNs
february_data = weather_data_pd[weather_data_pd['month'] == 2]
q_10 = february_data['humid'].mean() if not february_data['humid'].isna().all() else None
print("Mean humidity in February:", q_10)


# Question 11 What was the std for humidity in February? Float
print()
print('11' ,'output')
# Calculate the standard deviation of 'humidity' in February, ignoring NaNs
q_11 = february_data['humid'].std() if not february_data['humid'].isna().all() else None
print("Standard deviation of humidity in February:", q_11)



```
